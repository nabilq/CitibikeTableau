```python
import pandas as pd
```

```python
data = pd.read_csv("JC-201912-citibike-tripdata.csv")
```

```python
data
```

```python
#Start station ID
startstation= data['start station id'].value_counts()
startstation

#convert to DataFrame
startstation_df=pd.DataFrame(startstation)
startstation_df
```

```python


#set name of index
startstation_df.index.name='station id'

startstation_df
```

```python
#end station count
endstation=  data['end station id'].value_counts()
endstation
#convert to DataFrame
endstation_df=pd.DataFrame(endstation)

endstation_df
```

```python
#merge both DataFrames

endstation_df.columns=['Trips']

endstation_df
```

```python
#set name of index
endstation_df.index.name='station id'

endstation_df
```

```python
#merge both DataFrames
df_outer = pd.merge(startstation, endstation, on='station id', how='outer')

df_outer
```

```python
#add both columns
x= df_outer.sum(axis=1)
x
```

```python
#add series as new DataFrame
df_outer['Total Count']= x

df_outer
```

```python
#save DataFrame to new index

df_outer.to_csv('totalstationcount.csv')
```
